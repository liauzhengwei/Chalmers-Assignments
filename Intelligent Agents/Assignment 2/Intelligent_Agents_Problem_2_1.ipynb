Format the test set in two ways:
1. test set with answer
2. test set without answer for zero shot prompting

```python
def format_mcq_file(input_file, output_file, output_file_no_answer):
  with open(input_file, 'r') as infile, \
        open(output_file, 'w') as outfile, \
        open(output_file_no_answer, 'w') as outfile_no_ans:
    for line in infile:
      # Split the input into question, options, and answer
      parts = line.strip().split('\t')

      # Extract question and options
      question = parts[0]
      options = parts[1:-1]
      answer = parts[-1].strip()

      # Format options with letters
      formatted_options = [f"{chr(65 + i)}: {opt}" for i, opt in enumerate(options)]

      # Join the formatted question and options
      formatted_question = f"{question} {' '.join(formatted_options)}"

      # Save to file
      outfile.write(formatted_question + "\n")
      outfile.write(answer + "\n")

      outfile_no_ans.write(formatted_question + "\n")
      outfile_no_ans.write("\n")

def print_mcq_file(file_path, num_lines = 10):
  with open(file_path, 'r') as file:
    for _ in range(num_lines):
      print(file.readline().strip())

# Example usage
input_file = "/content/TestSet.txt"
output_file = "mcq_output_with_answer.txt"
output_file_no_answer = "mcq_output_no_answer.txt"
format_mcq_file(input_file, output_file, output_file_no_answer)
print_mcq_file(output_file)
print("\n")
print_mcq_file(output_file_no_answer)
```

##Few Shot Prompting
- User provides guidance in the form of a few specific examples
- Augment the data in a separate file with one or a few examples where you also give the correct answer
  - Begin each prompt with "The following are multiple choice questions (with answers) about [subject]
  - Then add up to 5 demostration examples with answers to the prompt before appending the question.
  - All prompts end with "Answer"

### Need to format the additional data for few shot prompting also

```python
def format_additional_data_file(input_file, output_file, output_file_no_answer):
  with open(input_file, 'r') as infile, \
        open(output_file, 'w') as outfile:
    for line in infile:
      # Split the input into question, options, and answer
      parts = line.strip().split('\t')

      # Extract question and options
      question = parts[0]
      options = parts[1:-1]
      answer = parts[-1].strip()

      # Format options with letters
      formatted_options = [f"{chr(65 + i)}: {opt}" for i, opt in enumerate(options)]

      # Join the formatted question and options
      formatted_question = f"{question} {' '.join(formatted_options)}"

      # Save to file
      outfile.write(formatted_question + " " + answer + "\n")

def print_additional_data_file(file_path, num_lines = 10):
  with open(file_path, 'r') as file:
    for _ in range(num_lines):
      print(file.readline().strip())

# Example usage
input_file = "/content/AdditionalDataForFewShotPrompting.txt"
output_file = "additional_data_with_answer.txt"
format_additional_data_file(input_file, output_file, output_file_no_answer)
print_additional_data_file(output_file)
```

```python
import random

# Read all available questions from a text file
with open("additional_data_with_answer.txt", "r", encoding="utf-8") as file:
    all_questions = file.readlines()

# Read the main questions
with open("mcq_output_no_answer.txt", "r", encoding="utf-8") as file:
    questions = file.readlines()

# Open the output file
with open("Llama 2 shot answers.txt", "w", encoding="utf-8") as output_file:
  for question in questions:
    if question.strip():
      # Select two random questions
      random_questions = random.sample(all_questions, 2)

      # Prepare the introduction text
      intro_text = "The following are multiple choice questions with answers about astronomy.\n\n"
      random_question_text = "".join(random_questions) + "\n"

      # Generate response
      full_prompt = intro_text + random_question_text + f"Question: {question.strip()}\nAnswer:"
      response = model.generate(full_prompt, max_tokens=25)

      output_text = f"{intro_text}{random_question_text}Question: {question.strip()}\n{response}\n"

      # Write to file
      output_file.write(output_text)

      # Print to console
      print(output_text)

print("Answers saved to Llama 2 shot answers.txt")
```

###Analysing results

```python
# Read the correct answers
with open("mcq_output_with_answer.txt", "r") as file:
  lines = file.readlines()
  correct_answers = [lines[i].strip() for i in range(1, len(lines), 2)]

# Read predicted answers
with open("Llama 0 shot answers edited.txt", "r") as file:
  lines = file.readlines()
  llama_0_shot_predictions = [lines[i].strip() for i in range(1, len(lines), 2)]

with open("Phi3 0 shot answers edited.txt", "r") as file:
  lines = file.readlines()
  phi3_0_shot_predictions = [lines[i].strip() for i in range(1, len(lines), 2)]

# Ensure both lists are the same length
if len(correct_answers) != len(llama_0_shot_predictions) or len(correct_answers) != len(phi3_0_shot_predictions):
  raise ValueError("Lists must have the same length")

print("Correct Answers:", correct_answers)
print("Llama 0 Shot Predictions:", llama_0_shot_predictions)
print("Phi3 0 Shot Predictions:", phi3_0_shot_predictions)

# Compare answers
correct_llama_0_shot = sum(1 for a, p in zip(correct_answers, llama_0_shot_predictions) if a == p)
correct_phi3_0_shot = sum(1 for a, p in zip(correct_answers, phi3_0_shot_predictions) if a == p)

print("Correct Llama 0 Shot Predictions:", correct_llama_0_shot)
print("Correct Phi3 0 Shot Predictions:", correct_phi3_0_shot)

total_questions = len(correct_answers)
accuracy_llama_0_shot = int((correct_llama_0_shot / total_questions) * 100)
accuracy_phi3_0_shot = int((correct_phi3_0_shot / total_questions) * 100)

print(f"Accuracy Llama 0 Shot: {accuracy_llama_0_shot} %")
print(f"Accuracy Phi3 0 Shot: {accuracy_phi3_0_shot} %")
```
