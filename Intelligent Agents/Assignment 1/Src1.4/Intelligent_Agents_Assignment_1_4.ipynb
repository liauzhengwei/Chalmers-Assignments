```python
import nltk
import wikipedia
import time
import re
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize,sent_tokenize
from google.colab import drive
drive.mount('/content/drive')
```

```python
# Output file
output_file = "/content/drive/My Drive/wikipedia_sentences.txt"

# Token count target
target_tokens = 5000000
#target_tokens = 1000

total_tokens = 0

#Regex pattern to filter non-English characters
english_pattern = re.compile(r'^[A-Za-z0-9.,!?;:\'"\s-]+$')

with open(output_file, "w", encoding="utf=8") as out_f:
    while total_tokens < target_tokens:
      try:
          title = wikipedia.random() # Get a random Wikipedia article title
          page = wikipedia.page(title) # Fetch the article
          text = page.content # Extract content
          print(f"Processing: {title}")

          # Extract full sentences
          sentences = sent_tokenize(text)

          for sentence in sentences:
            tokens = word_tokenize(sentence) # Tokenize

            # Filter out section titles
            if "==" in sentence:
              continue

            # Filter out equations
            if any(char in sentence for char in "=+-*/<>^_"):
              continue

            # Filter out non english patterns
            if not english_pattern.match(sentence):
              continue

            # Filter out very short sentences
            if len(tokens) <= 5:
              continue

            total_tokens += len(tokens)
            out_f.write(sentence + "\n") # Save to file
            out_f.flush() # Flush to ensure data is written immediately
            print(f"Added {len(tokens)} tokens. Total tokens: {total_tokens}/{target_tokens}")

      except Exception as e:
        print(f"Error: {e}")
        time.sleep(1)

print("Done")
```
